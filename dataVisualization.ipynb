### **Initial Loading of the Data**

Now that the data has been pre-processed let's do a little exploratory data analysis on said data. Let's first get the summary stats of each column in the dataset:


```python
import pandas as pd

data = pd.read_csv('C:\\Users\\Mike Delevan\\PycharmProjects\\Senior-Project\\data-scraper\\combinedData.csv')
data.drop('Unnamed: 0', axis=1,inplace=True)

data.describe()
```

Let's plot a scatter matrix to get plots for the relationships between all variables. We can then take this plot to find interesting correlations between variables and then explore them more **(Note: This may take awhile to run)**:


```python
%matplotlib inline
from pandas.plotting import scatter_matrix
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

scatter_matrix(data, figsize=(24,16))
plt.show()
```

From this graph we can see that many of the variables are positively correlated with one another. Lets check out the relationship between the win column and OPS column

```python
import seaborn as sns
sns.boxplot(x=data["Win"], y=data["OPS"] )
```

Based on this graph, one could say that the higher a team's OPS is, the greater chance they have of winning a game. It may be a good indicator of a win when fed into a machine learning algorithm.

Let's see the relationship between the number of strikeouts and a win:

```python
import seaborn as sns
sns.boxplot(x=data["Win"], y=data["Strikeouts"] )
```
