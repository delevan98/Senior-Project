### **Visualization and Summarization of the Data**

Now that the data has been pre-processed let's do a little exploratory data analysis on said data. I realized I included this in the data processing notebook, therefore I moved all of that code and graphs over to the right file. Some of the graphs in here will look like the ones in data processing however I'll be using the combined data file to make graphs (~40,000 rows).

The goals of this exploratory data analysis include:

* Determining variables that may be good indicators of a win
* Determining how those variables differentiate teams
* Validating our findings using a correlation matrix

Before we do anything we must first get the summary stats of each column in the dataset:


```python
import pandas as pd

data = pd.read_csv('C:\\Users\\Mike Delevan\\PycharmProjects\\Senior-Project\\data-scraper\\combinedData.csv')
data.drop('Unnamed: 0', axis=1,inplace=True)

data.describe()
```

We can get a lot of useful information regarding the data using the output from describe. One important finding is tha in the dataset there is approximately 50% rows with a win value and approximately 50% rows with a loss value. This is important for machine learning because we want to give the model evenly distributed data. Just having an even split might even reduce overfitting because the model will be generalized to both win and loss thus being a better predictor. If there were more wins in the dataset, the system would learn from that data and may end up predicting more wins than losses, which is not the intended function of the system.

We can also see that on average, teams score around 4 points a game which may seem low, but compared to soccer it is a pretty good value. Also, on average a team can attribute at least one run to a homerun and the other 3 points through batting base runners in. These kinds of relationships can be better seen using graphs so let's do just that

Let's plot a scatter matrix to get plots for the relationships between all variables. We can then take this plot to find interesting correlations between variables and then explore them more **(Note: This may take a while to run)**:


```python
%matplotlib inline
from pandas.plotting import scatter_matrix
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

scatter_matrix(data, figsize=(24,16))
plt.show()
```

From this graph we can see that many of the variables are positively correlated with one another. However it's hard to see what variables are correlated with win because it's value is either 0 or 1. We can enlarge one of win graphs to better see the values and how they affect the Win value. So, let's check out the relationship between the win column and OPS column:

```python
import seaborn as sns
sns.boxplot(x=data["Win"], y=data["OPS"] )
```

Based on this graph, one could say that the higher a team's OPS is, the greater chance they have of winning a game. It may be a good indicator of a win when fed into a machine learning algorithm. Obviously, having this graph isn't enough to conclude that the differences in OPS values is significant or in anyway correlated with a win. Let's calculate the Pearson's correlation coefficient to try and validate our findings from this graph:


```python
print(data["Win"].corr(data["OPS"]).round(3))
```

The Pearson's correlation coefficient values can range from +1 to -1, where +1 indicates a perfect positive correlation while -1 indicates a perfect negative correlation. Looking at our correlation coefficient, we can see that it agrees with our previous assumption that a team's OPS is a high indicator of a win. It's not a direct correlation, but there is some kind of relationship between OPS and Win. If time permits, a statistical analysis can be done to determine whether the differences in OPS values for win/loss is significant.

Let's try a different varaible and see whether it correlates with a win or not. Here we will be using offensive striketouts:

```python
sns.boxplot(x=data["Win"], y=data["Strikeouts"] )
```

It seems that the averages strikeouts for both win and loss are rather close. It seems from the graph that strikeouts are negatively correlated to win/loss meaning that having more offensive strikeouts could result in a loss. This graph makes me believe that stikeouts are not a very strong predictor of a win/loss because of the narrow difference of the mean.

Let's validate our assumption through finding the Pearson correlation between win/loss and strikeouts:

```python
print(data["Win"].corr(data["Strikeouts"]).round(3))
```

Our Pearson correlation value confirms our previous assumption that strikeouts are not a strong indicator of a win/loss. In fact, it tells us that there is a slight negative correlation which is shown in the graph above.

Now, we could sit here all day and do this for every variable in the dataset, but that will get boring after about the third variable we explore. That is why we will try to find the relationship between win/loss and one more variable to keep things short. Let's explore the score variable and it's correlation with Win:

```python
sns.boxplot(x=data["Win"], y=data["Score"])
```

Scores in baseball games can be extremely variable where a game could end 1-0, 15-14, or 20-1. Using this boxplot one can see that typically when a team wins, they have a higher score than when they lose. Even if a team loses there's is still a chance that they will have a high score as well as the winning team. However, from the graph we can infer that the score could be a very strong predictor of a win or loss. If a team scores a lot of points, there's a very good chance that they will win the game and if they score a small amount of points there's a good chance they will lose the game. This variable may actually have a direct correlation with the win/loss variable, however I am unsure of this and will research it more as the project progresses.

Let's confirm our assumption that score is a good predictor of a win/loss:


```python
print(data["Win"].corr(data["Score"]).round(3))
```

From this value we can see that the score variable has a high correlation with win/loss. Compared with OPS the score variable is slightly more correlated with a .558 correlation value versus a .524 correlation value.

Let's now step away from doing a bunch of boxplots and see how a team's differe based on their average OPS over the 2018 season:

```python
import numpy as np
import matplotlib.pyplot as plt

team1 = pd.read_csv('C:\\Users\\Mike Delevan\\PycharmProjects\\Senior-Project\\data-scraper\\BOS_Full.csv')
team1Copy = pd.DataFrame(columns=['teamAbbr', 'League', 'Score', 'isHomeTeam', 'atBats', 'Hits',
                                      'Doubles', 'Triples', 'homeRuns', 'RBI', 'Walks', 'Strikeouts', 'LOB',
                                      'pitchersUsed', 'indER', 'teamER', 'Errors', 'battingAverage', 'OBP', 'Slugging',
                                      'OPS', 'Win', 'OPSAverage'])

team1size = team1.shape[0]
y=0
for x in range(team1size - 162, team1size):
    
    team1Copy.loc[y+1] = team1.loc[x]
    y = y+1

avgArray = np.array([])
sum = 0
for (idx, row) in team1Copy.iterrows():
    sum = sum + row['OPS']
    average  = sum / (len(avgArray)+1)
    avgArray = np.append(avgArray, average)

team1Copy['OPSAverage'] = avgArray



team2 = pd.read_csv('C:\\Users\\Mike Delevan\\PycharmProjects\\Senior-Project\\data-scraper\\NYN_Full.csv')
team2Copy = pd.DataFrame(columns=['teamAbbr', 'League', 'Score', 'isHomeTeam', 'atBats', 'Hits',
                                      'Doubles', 'Triples', 'homeRuns', 'RBI', 'Walks', 'Strikeouts', 'LOB',
                                      'pitchersUsed', 'indER', 'teamER', 'Errors', 'battingAverage', 'OBP', 'Slugging',
                                      'OPS', 'Win', 'OPSAverage'])
team2size = team2.shape[0]
y=0
for x in range(team2size - 162, team2size):
    team2Copy.loc[y+1] = team2.loc[x]
    y = y+1

avgArray = np.array([])
sum = 0
for (idx, row) in team2Copy.iterrows():
    sum = sum + row['OPS']
    average  = sum / (len(avgArray)+1)
    avgArray = np.append(avgArray, average)

team2Copy['OPSAverage'] = avgArray



combined = pd.DataFrame(columns=['teamAbbr', 'League', 'Score', 'isHomeTeam', 'atBats', 'Hits',
                                      'Doubles', 'Triples', 'homeRuns', 'RBI', 'Walks', 'Strikeouts', 'LOB',
                                      'pitchersUsed', 'indER', 'teamER', 'Errors', 'battingAverage', 'OBP', 'Slugging',
                                      'OPS', 'Win', 'OPSAverage'])

combined = combined.append(team1Copy)
combined = combined.append(team2Copy)
color = ['blue', 'gray']
combined.groupby("teamAbbr")['OPSAverage'].plot(figsize=(24,7), legend=True, title='Average OPS during 2018 Season')
plt.xlabel('Game Number')
plt.ylabel('Average OPS')
plt.show()
```

From this graph we can see that when the season starts the average OPS for both teams is all over the place. But, over the course of the season each team settles into an average OPS value. In this case the Boston Red Sox have a higher average OPS value than the New York Mets. Using our previous finding that OPS is a good predictor of a win/loss, we could infer that the Red Sox would win more often than the Mets. Checking the [Baseball-Reference](https://www.baseball-reference.com/leagues/MLB/2018-standings.shtml) standings, we see that the Red Sox won a humongous 108 games during the 2018 season versus the Mets' measly 77 wins. Again this confirms our inference that a higher OPS can lead to a better win percentage. Hopefully the machine learning model will pick up on this relationship to increase it's accuracy.

Because the Red Sox won the World Series in the 2018 season, I'd like to quickly see how their OBP value compared with every other average OBP value during that season:

```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import islice

teamAbbr = ["CHN","PHI","PIT", "CIN", "SLN", "BOS", "CHA",
            "CLE", "DET", "NYA", "BAL", "LAN", "SFN", "MIN",
            "HOU", "NYN", "ATL", "OAK", "KCA", "SDN", "TEX",
            "TOR", "SEA", "FLO", "COL", "ANA", "TBA", "ARI",
            "MIL", "WAS"]

avgArray = np.array([])
for x in range(30):
    team1 = pd.read_csv('C:\\Users\\Mike Delevan\\PycharmProjects\\Senior-Project\\data-scraper\\'+teamAbbr[x]+'_Full.csv')
    team1size = team1.shape[0]
    sum = 0
    average = 0

    for (idx, row) in islice(team1.iterrows(), team1size-162, None):
        sum = sum + row['OBP']
   
    average  = sum / 162
    avgArray = np.append(avgArray, average)
    
plt.hist(avgArray, color='grey', lw=1)
plt.xlabel('Average OBP')
plt.ylabel('Count')
plt.title('Average OBP Values in the 2018 Season')
plt.axvline(x=0.3319, color='crimson', label='Boston Avg OBP')
plt.axvline(x=0.3225, color='dodgerblue', label='Dodgers Avg OBP')
plt.legend()
```

Let's move onto the correlation heatmap to get a visualization of our assumptions and findings from our analysis:

```python
import seaborn as sns
data = pd.read_csv('C:\\Users\\Mike Delevan\\PycharmProjects\\Senior-Project\\data-scraper\\combinedData.csv')

data.drop(['teamAbbr','League', 'Unnamed: 0'], axis=1, inplace=True)

plt.subplots(figsize=(35,35))
cor = data.corr()
sns.heatmap(cor, cmap=plt.cm.Reds)
plt.gcf().subplots_adjust(top=.95,bottom=0.35)
plt.xticks(rotation=45,ha='right')
fig = plt.gcf()
plt.show()
```

In our previous analysis we assumed that score and OPS were good predictors of a win/loss and strikeouts were not a good indicator of a win/loss. Using this correlation heatmap we can see that score, hits, RBI, BA, OBP, slugging, and OPS are strongly correlated with win/loss which agrees with our findings. Strikeouts are shown as being minimally correlated with a win/loss which does agree with our assumption that strikeouts don't strongly indicate a win/loss. However, it seems the earned runs values are not correlated at all with a win/loss which seems counter intuitive. This could be because there was no feature engineering done for that variable so it doesn't really correlate with win/loss in its current form. If earned runs average is calculated and included into the dataset, these values may find new purpose later in the project.
